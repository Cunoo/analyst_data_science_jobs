# Data cleaning

```python
df.drop(["Unnamed: 0"], inplace=True, axis=1)
```

```python
df
```

```python
df['job_title'] = df['job_title'].replace('ML Engineer', 'Machine Learning Engineer')
df.head(49)
```

```python
df.dtypes
```

# Data visualization

```python
job_title_data = df['job_title'].sort_values()


plt.figure(figsize=(10,9)) #acclimatize size
sns.histplot(y=job_title_data, color="skyblue", bins=1).set_title('popular roles')
#plt.xticks(rotation='vertical')
```

```python
df['job_title'].value_counts()
```

```python
plt.figure(figsize=(10,9)) #acclimatize size

    
data_scientist_df = df[df['job_title'] == 'Data Scientist']
sns.boxplot(data=data_scientist_df ,x= 'experience_level', y = 'salary_in_usd',).set_title('Data Scientist Salary')
```

```python
plt.figure(figsize=(10,10)) #acclimatize size
#plt.xticks(rotation='vertical')
sns.scatterplot(data=df['company_location']).set_title('where are located most company')
```

```python
comp = df[df['company_location'] == 'US']

sns.boxplot(data=comp, x='company_size', y = 'salary_in_usd') 
```

```python
data =  df
plt.figure(figsize=(10,15)) #acclimatize size
sns.boxplot(data=df, y='job_title', x = 'salary_in_usd').set_title('salary in job titles')
```

```python
df_titles= df[df['job_title'] == 'Data Analyst']
#how salary grow up from 2020 to 2022
sns.lineplot(data=df_titles, x='work_year', y='salary_in_usd').set_title('how salary grow up in Data Analyst')
plt.xticks(rotation='vertical')
```
